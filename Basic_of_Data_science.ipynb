#Data wrangling, Cleansing

```python
df = pd.read_csv(base_repo + '2019_Yellow_Taxi_Trip_Data.csv')
```

```python
df.head(5)
```

```python
df.columns.to_list()
```

```python
mask = df.columns[df.columns.str.contains('id$|store_and_fwd_flag')]
```

```python
df[mask]
```

```python
df_clean = df.drop(columns = mask)
```

```python
df_clean.head()
```

```python
df_clean = df_clean.rename(columns = {
    'tpep_pickup_datetime' : 'pickup_dt',
    'tpep_dropoff_datetime' : 'dropoff_dt'
})
df_clean
```

# change dtype

```python
df_clean.dtypes
```

```python
sel_cols = ['pickup_dt','dropoff_dt']
df_clean[sel_cols] = df_clean[sel_cols].apply(pd.to_datetime)
```

```python
df_clean.dtypes
```

```python
df_clean['pickup_dt'].dt.hour
```

```python
df_clean['pickup_dt'].dt.year
```

#feature engineering

```python
df_clean['trip_time'] = df_clean['dropoff_dt'] - df_clean['pickup_dt']
df_clean
```

```python
df_clean['avg_speed'] = df_clean['trip_distance'] / df_clean['trip_time'].dt.total_seconds() / 60 / 60
```

```python
df_clean.head()
```

#Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(base_repo + 'tsa_melted_holiday_travel.csv')
df.head()
```

```python
df.dtypes
```

```python
df['date'] = df['date'].map(pd.to_datetime)
df.dtypes
```

```python
df.plot(x='date',y='travelers', figsize=(12,6))
```

```python
df.head(5)
```

```python
plot_df = df.pivot_table(index='date',values='travelers',columns='year')
plot_df
```

```python
plot_df.plot(figsize = (16,8))
```

```python
plot_df[2020].dropna().plot(figsize = (16,8),
                     title='travellers',
                     xlabel = 'date',
                     ylabel='traveller',
                     marker='.',
                     color='g')
```

```python
plot_df[2020].dropna().resample('M').sum().plot(figsize=(16,8))
```

```python
plot_df[2020].dropna().resample('W').sum().plot(figsize=(16,8))
```

## bar plot

```python
df_p = df.copy()
df_p['month'] = df_p['date'].dt.month
df_p
```

```python
df_p = df_p.pivot_table(index='month',
                 columns = 'year',
                 values='travelers')
```

```python
sns.heatmap(df_p, cmap ='Blues', annot=True, fmt = '.10g')
```

```python
ax = df_p.plot(kind='bar', ylabel ='travelers',figsize=(16,8))

ax.legend(title='',loc='center',bbox_to_anchor=(0.5,-0.15),ncol = 3,frameon = False)
```

##plot distributions

```python
df.head()
```

```python
df.query('year == 2019').plot(
    kind='hist',
    x='date',
    y='travelers',
    bins=30,
    figsize=(16,8)
)
```

```python
pivot_df = df.pivot_table(columns='year',
               values='travelers',
               index='date')
pivot_df
```

```python
fig,ax = plt.subplots(3,1,figsize=(16,8))
ax = pivot_df.plot(
    kind='hist',
    subplots=True,
    bins=30,
    figsize=(16,8),
    ax = ax,
    sharex=True,
    density=True
)
pivot_df.plot(
    kind='kde',
    subplots=True,
    ax=ax,
    color='red'
)
```

```python
pivot_df.plot(
    kind='box'
)
```
